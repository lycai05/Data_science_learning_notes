```python
import pandas as pd
import numpy as np
import copy

def SWBackTrack(v, w):
    
    # index for s
    v_list = list(v)
    v_index = [v_list[i] + "_" + str(i + 1) for i in range(len(v_list))]
    v_index.insert(0, "v_Gap_0")
    
    # column names for s
    w_list = list(w)
    w_index = [w_list[i] + "_" + str(i + 1) for i in range(len(w_list))]
    w_index.insert(0, "w_Gap_0")
    
    s = pd.DataFrame(index=v_index, columns=w_index, data=-np.inf)
    #print(s)
    
    # initialize penalty score
    s.iloc[:,0] = [0] * (len(w_list) + 1) 
    s.iloc[0,:] = [0] * (len(v_list) + 1)
    #s.iloc[0,0] = 0
    
    backTrack = copy.deepcopy(s)
    #print(s)
    
    def delta_score(m, n):
        if m == n:
            return 2
        else:
            return -1
    
    # recursion
    for i in range(1, len(v)+1):
        for j in range(1, len(w)+1):
            s.iloc[i, j] = max(s.iloc[i-1, j-1] + delta_score(v[i - 1], w[j - 1]), s.iloc[i-1,j] - 1, s.iloc[i, j-1] - 1, 0)
            if s.iloc[i, j] == s.iloc[i-1, j-1] + delta_score(v[i - 1], w[j - 1]):
                backTrack.iloc[i, j] = "\\"
            elif s.iloc[i, j] == s.iloc[i-1,j] - 1:
                backTrack.iloc[i, j] = "|"  
            elif s.iloc[i, j] == s.iloc[i,j-1] - 1:
                backTrack.iloc[i, j] = "-"
            else:
                backTrack.iloc[i, j] = "0"
    print(s.stack().index[np.argmax(s.values)])
    return backTrack
                                 

def OutputSW(backTrack, v, w):
    v_align = []
    w_align = []
    #df.loc[df['favcount'].idxmax(), 'sn']
    i = len(v)
    #print(i)
    j = len(w)
    #print(j)
    
    while i != 0 and j != 0:
        if backTrack.iloc[i, j] == "\\":
            v_align.insert(0, v[i - 1])
            w_align.insert(0, w[j - 1])
            i = i - 1
            j = j - 1
        elif backTrack.iloc[i, j] == "|":
            v_align.insert(0, v[i - 1])
            w_align.insert(0, "_")
            i = i - 1
        elif backTrack.iloc[i, j] == "-":
            v_align.insert(0, "_")
            w_align.insert(0, w[j - 1])
            j = j - 1
        else:
            break

    return v_align, w_align
```
