#### 梯度下降的概念
一維度的純量x的梯度就是算f(x)對x的微分（一个值），多維度的向量x的梯度就是算f(x)對x所有元素的偏微分（一个向量）
1. 用途：用于优化目标函数
2. 途径：通过最小化损失函数

#### 梯度下降法的学习率（learning rate） $\eta$
1. 是一个超参数
2. 影响获得最优解的速度
3. 影响是否会取得最优解

#### 梯度下降法的极值点
1. 并非所有的损失函数都有唯一的极值点，
2. 存在局部最优解，而并非全局最优解
#### 解决方案：
1. 多次运行，随机化初始点
2. 梯度下降的初始点也是一个超参数

```python
# 使用梯度下降最优化简单一元线性回归的损失函数
import numpy as np
import matplotlib.pyplot as plt

plot_x = np.linspace(-1, 6, 100)
plot_y = (plot_x - 1.5)**2 -1
plt.plot(plot_x, plot_y)
plt.show()
```

```python
# 设定学习率为0.01
# 设定迭代次数为100
# 通过计算微分的正负来确定梯度下降的方向，从而确定x更新的方向
#i = 0
def gradient_descent(theta, learning_rate):
    theta = theta
    learning_rate = learning_rate
    theta_arr = [theta]
    while True:
        # 微分/梯度
        gradient = 2 * (theta - 1.5)
        # 储存当前的theta值
        last_theta = theta
        # 更新theta
        theta = theta - learning_rate * gradient
        theta_arr.append(theta)
        if abs(((theta - 1.5)**2 -1) - ((last_theta - 1.5)**2 -1)) < 0.00000001:
            break
       # i = i + 1
        #print("current theta is:")
        #print(theta)
        #print("current gradient is:")
        #print(gradient)  
    return theta_arr
        
        
#### 可视化梯度的更新
def plot_gradient_descent(theta_arr):
    plt.plot(plot_x, plot_y)
    plt.plot(np.array(theta_arr), (np.array(theta_arr) - 1.5)**2 -1, color = "r", marker = "+")
    plt.show()
```

```python
plot_gradient_descent(gradient_descent(-1, 0.01))
```

```python
plot_gradient_descent(gradient_descent(-1, 0.1))
```

```python
plot_gradient_descent(gradient_descent(-1, 0.5))
```

```python
可以看到不同的学习率对梯度下降的影响很大，但学习率太大时，有可能直接跳过了最小值
```
