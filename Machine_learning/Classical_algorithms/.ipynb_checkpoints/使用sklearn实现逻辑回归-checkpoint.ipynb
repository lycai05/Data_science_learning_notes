```python
import pandas as pd
colNames = ['Sample code number','Clump Thickness', 'Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion', 'Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data",
                  names = colNames)
data.head(10)
```

```python
import numpy as np
data = data.replace(to_replace = "?", value = np.nan)
data = data.dropna()
print(data)
x = data[colNames[1:10]]
y = data[colNames[10]]
```

```python
def LogisticReg(x, y):
    
    # split data into training and test
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
    
    # initialize linearRegression
    from sklearn.linear_model import LogisticRegression
    lg = LogisticRegression()
    lg.fit(x_train, y_train)
    print(lg)
    print(lg.predict(x_test))
    
    #
    from sklearn.metrics import classification_report
    print(classification_report(lg.predict(x_test), y_test))
    
    return None


if __name__ == "__main__":
    LogisticReg(x, y)
```
