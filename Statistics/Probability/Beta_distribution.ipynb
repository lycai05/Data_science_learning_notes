```r
### Beta distribution

set.seed(666)

gamma <- function(x) {
    return(factorial(x-1))
}


beta_function <- function(x, alpha, beta) {
    return(gamma(alpha+beta)/(gamma(alpha)*gamma(beta)) * x^(alpha-1) * (1-x)^(beta-1))
}

x <- seq(0, 1, 0.001)
alpha <- 1
beta <- 3

par(mfrow=c(2,2))
for (i in list(c(1, 3), c(5,1), c(2, 2), c(2, 5))) {
    p <- plot(x, beta_function(x, i[1], i[2]))
}
```

```r
beta(x, alpha, beta)
```
