```r
### benoulli distribution

set.seed(666)
bernoulli <- function(k, p) {
    y <- sapply(k, function(m){if(m) {
        return(p)
        } else {
            return(1-p)
        }
    })
    return(y)
}

n <- 100
p <- 0.6

plot(1:n, bernoulli(sample(c(TRUE,FALSE), n, TRUE),0.6))
```
