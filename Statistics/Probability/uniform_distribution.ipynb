```r
### uniform distribution
set.seed(666)
# x is a vector
# [a, b] is the range in which x = sth
uniform <- function(x, a, b) {
    y <- sapply(x, function(m) {
        if (m >=a & m <= b) {
            return(1/(b-a))
        } else {
            return(0)
        }
    })
    print(mean(y))
    print(sd(y))
    return(y)
}

x <- seq(-100, 100)
plot(x, uniform(x, -50, 50))
```
