##### 单向循环链表与单链表的区别
1. 单链表尾节点指向空地址
2. 循环链表的尾节点指针指向链表的头节点
3. 与单向链表相比，优点是从尾节点到头节点比较方便

![循环链表](figures/循环链表.png)

```python
class Node(object):
    def __init__(self, elem):
        self.elem = elem
        self.next = None
        
        
class CircularSingleLinkList(object):
    def __init__(self, node = None):
        self._head = node
        cur = node
        while cur.next != None:
            cur = cur.next
        cur.next = node
        
        
    # 判断是否为空
    def is_empty(self):
        return self._head == None
    
    
    # 计算长度
    def length(self):
        if self.is_empty():
            count = 0
        else:            
            count = 1
            cur = self._head
            while cur.next != self._head:
                cur = cur.next
                count += 1
        return count
        
    # 遍历
    def travel(self):
        if self.is_empty():
            print("This CSLL is empty.")
        else:            
            count = 1
            cur = self._head
            print("The {} element is: {}".format(count, cur.elem))
            while cur.next != self._head:
                cur = cur.next
                count += 1
                print("The {} element is: {}".format(count, cur.elem))
        return None
    
    # 在头部添加元素
    def prepend(self, elem):
        node = Node(elem)
        print("Prepend {} to the CSLL".format(node.elem))
        if self.is_empty():
            self._head = node
            node.next = node
        else:
            cur = self._head
            while cur.next != self._head:
                cur = cur.next
            cur.next = node
            node.next = self._head
            self._head = node
                        
        
    # 在尾部添加元素    
    def append(self, elem):
        node = Node(elem)
        print("Append {} to the CSLL".format(node.elem))
        if self.is_empty():
            self._head = node
            node.next = node
        else:
            cur = self._head
            while cur.next != self._head:
                cur = cur.next
            cur.next = node
            node.next = self._head
    
    
    # 删除元素，根据元素的内容，删除遇到的第一个
    def delete(self, elem):
        pass
#        if self.is_empty():
#            print("The CSLL is empty. Nothing to delete")
#        else:
#            if self._head.elem = elem:
                
               
    # 在指定元素后插入    
    def insert(self, afterElem, insertElem):
        insertNode = Node(insertElem)
        print("Insert {} to the CSLL after {}".format(insertElem, afterElem))
        if self.is_empty():
            self._head = node
            node.next = node
            return
        else:
            cur = self._head
            while cur.next != self._head:
                if cur.elem == afterElem:
                    insertNode.next = cur.next
                    cur.next = insertNode
                    return
                else:
                    cur = cur.next
            if cur.elem == afterElem:
                insertNode.next = cur.next
                cur.next = insertNode
                return
            print("There is no {}".format(afterElem))
            return
    
    def search(self, elem):
        cur = self._head
        count = 1
        while cur.next != self._head:
            if cur.elem == elem:
                print("{} is the {} element".format(elem, count))
                return
            else:
                cur = cur.next
                count += 1
        if cur.elem == elem:
            print("{} is the {} element".format(elem, count))
            return
        print("There is no {}".format(elem))
        return
```
